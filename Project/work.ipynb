```python
import numpy as np
import cv2
from win10toast import ToastNotifier

import time
import datetime

import smtplib
from email.mime.text import MIMEText
from email.header import Header
from email.mime.multipart import MIMEMultipart
xml = 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(xml)

cap = cv2.VideoCapture(0) 
cap.set(3,640)  
cap.set(4,480) 
count=0
t=0

s = time.time()
c=0
wt=0
while(True):

    start = time.time()
    ret, frame = cap.read() 
    frame = cv2.flip(frame, 1) 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray,1.05,5) 
    print("Number of faces :" + str(len(faces)))
    
    cv2.imshow('result', frame)

    if len(faces)==0:
        t2=time.time() - start
        t+=t2        
        print(t)
        if round(t,0)==2:
            print("자리를 비운지 2분이 지났습니다 착석해주세요")      
            toaster = ToastNotifier()
            toaster.show_toast("empty","회의시간을 비운지 2 분이 지났습니다 자리에 착석 부탁드립니다")
            t+=1
        elif round(t,0)-1==5:
            print("자리를 비운지 5분이 지났습니다 착석해주세요")      
            toaster = ToastNotifier()
            toaster.show_toast("empty","회의시간을 비운지 5 분이 지났습니다 자리에 착석 부탁드립니다")
            t+=1
            


    else:
        wt2=time.time() - start
        wt+=wt2   
        if round(t,0)-1>5:
            count+=t
            t=0
        else:
            t=0

    
        
    k = cv2.waitKey(30) & 0xff
    if k == 27: 
        if len(faces)==0:
            if round(t,0)-1>5:
                count+=t
        break

e=time.time() - s 
cap.release()
cv2.destroyAllWindows()


work = datetime.timedelta(seconds=e)

day = time.strftime('%Y-%m-%d', time.localtime(time.time()))




if wt/(count+wt)*100 >75:
    sco='양호'
elif wt/(count+wt)*100 >45:
    sco='미흡'
else:
    sco='주의'

message ="""Subject: Working History
To: {recipient}
From: {sender}

안녕하세요 {name}님
"""+str(day)+"""기준 근무 상태 기록을 메일로 보냅니다.

총 근무 시간"""+str(datetime.timedelta(seconds=count+wt))+" 중 " +str(datetime.timedelta(seconds=wt))+"""동안 근무를 하신걸로 측정되었습니다.

{name}님의 근무 상태는""" +str(int(wt/(count+wt)*100))+"%로 "+sco+"""로 판단됩니다.

오늘 하루도 수고 하셨습니다."""


sender_email = 'hanky0712@gmail.com' 
receiver_email = 'hanky0712@gmail.com'       
login = 'hanky0712@gmail.com'
password = 'hky1234!'
with smtplib.SMTP_SSL("smtp.gmail.com") as server:
    server.login(login, password)
    server.sendmail(sender_email,
                    receiver_email,
                    message.format(name = "user1", sender = sender_email , 
                                   recipient= receiver_email 
                            
                                   
                                  ).encode("utf-8")
                    )
print('Sent') 
```
